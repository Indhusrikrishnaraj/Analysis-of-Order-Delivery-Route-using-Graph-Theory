# Network Building

```python
import networkx as nx
G = nx.DiGraph()
```

```python
import csv
with open('distances.csv') as network:
  data = csv.reader(network)
  for cities in data:
    G.add_edge(cities[0], cities[1], distance = cities[2])
```

```python
print('Number of nodes', len(G.nodes))
print('Number of edges', len(G.edges))
```

# Optimization

## Dijkstra's Shortest Path Algorithm

```python
from queue import PriorityQueue

# Implementing the standard Dijkstra's Algorithm
def dijkstra(graph, start, end):
  distances = {node: float('inf') for node in graph.nodes}
  distances[start] = 0
  queue = PriorityQueue()
  queue.put((0, start))
  previous = {node: None for node in graph.nodes}

  while not queue.empty():
    current_distance, current_node = queue.get()

    if current_node == end:
      path = []
      while current_node:
        path.append(current_node)
        current_node = previous[current_node]
      return distances[end], path[::-1]

    for neighbor in graph.neighbors(current_node):
      distance = current_distance + int(graph[current_node][neighbor]['distance'])

      if distance < distances[neighbor]:
        distances[neighbor] = distance
        previous[neighbor] = current_node
        queue.put((distance, neighbor))

  return distances[end], None
```

```python
src = 'Kilis'
dest = 'Yalova'

shortest_path_distance, shortest_path = dijkstra(G, src, dest)

if shortest_path:
  print(f"Shortest path distance between {src} and {dest}: {shortest_path_distance}")
  print("Shortest path:", shortest_path)
else:
  print(f"There is no path between {src} and {dest}")
```

```python
dijkstra_color = []
dijkstra_size = []

for i in G.nodes:
  if i in shortest_path:
    G.nodes[i]['color'] = "ORANGE"
    G.nodes[i]['node_size'] = 2000
  else:
    G.nodes[i]['color'] = "#A0CBE2"
    G.nodes[i]['node_size'] = 1000
  dijkstra_color.append(G.nodes[i]['color'])
  dijkstra_size.append(G.nodes[i]['node_size'])
```

```python
fig = plt.figure(figsize=(18,12))
nx.draw(G, with_labels=True, node_size=dijkstra_size, node_color=dijkstra_color)
```

## Comparison with Default Path Algorithm

```python
from collections import defaultdict, deque

def bfs(graph, start, end):
    queue = deque()
    visited = {}
    path = {}
    dist = 0

    # Initializion
    for node in graph:
        visited[node] = False
        path[node] = None

    visited[start] = True
    queue.append(start)

    # BFS
    while queue:
        # Dequeue a node
        current_node = queue.popleft()

        if current_node == end:
            shortest_path = []
            while current_node is not None:
                shortest_path.append(current_node)
                current_node = path[current_node]
            return shortest_path[::-1]

        for neighbor in graph[current_node]:
            if not visited[neighbor]:
                visited[neighbor] = True
                path[neighbor] = current_node
                queue.append(neighbor)

    # no path
    return None
```

```python
def total_weight(path):
  total_weight = 0
  for i in range(len(path)-1):
    total_weight += int(G[path[i]][path[i+1]]['distance'])
  return total_weight
```

```python
start_node = 'Kilis'
end_node = 'Yalova'
path = bfs(G, start_node, end_node)

print(f"Path from {start_node} to {end_node}: {' -> '.join(path)}")

total_weight = total_weight(path)
print("Total weight of the path:", total_weight)
```
